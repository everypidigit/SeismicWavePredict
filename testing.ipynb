# Loading up the data, checking it, making some preprocessing

```python
import pandas as pd

df = pd.read_csv("./Archive/chan2.csv")

for col in df.columns:
    df[col] = [float(str(val).replace(' ','').replace(',','.')) for val in df[col].values]

df = df.astype(float)
df.dtypes
```

```python
df
```

```python
def min_max_scaler(df):
  min_values = df.min(axis=0)
  max_values = df.max(axis=0)
  scaled_df = (df - min_values) / (max_values - min_values)
  return scaled_df

df_scaled_minmax = min_max_scaler(df.copy())
df_scaled_minmax = df_scaled_minmax.dropna(axis=1)
df_scaled_minmax
```
